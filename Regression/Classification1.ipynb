```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def hypothesis(x,w,b):
    '''accepts one input vector c , input weight vector w and bias b'''
    h = np.dot(x,w) + b
    return sigmoid(h)
     
def sigmoid(z):
    return 1.0 / (1.0  + np.exp(-1.0*z))

def error(y_true,x,w,b):
    m = x.shape[0]
    err = 0.0
    for i in range (m):
        hx = hypothesis(x[i],w,b)
        err += y_true[i]*np.log2(hx) + (1-y_true[i])*np.log2(1-hx)
    return -err/m

def get_grad(y_true,x,w,b):
    grad_w = np.zeros(w.shape)
    grad_b = 0.0
    m = x.shape[0]
    for i in range (m):
        hx = hypothesis(x[i],w,b)
        grad_w += (y_true[i]-hx)*x[i]
        grad_b += (y_true[i]-hx)
        
    grad_w /= m
    grad_b /= m
    
    return [grad_w ,grad_b]
    
def grad_descent(x,y_true,w,b,learning_rate = 0.1):
    '''one iteration of gradient descent'''
    err = error(y_true,x,w,b)
    [grad_w, grad_b] = get_grad(y_true,x,w,b)
    
    w = w + learning_rate*grad_w
    b = b + learning_rate*grad_b
    
    return err,w,b

def predict(x,w,b):
    confidence = hypothesis(x,w,b)
    if (confidence < 0.5):
        return 0
    else:
        return 1
    
def get_acc(x_test,y_test,w,b):
    y_pred = []
    sum1 = 0.0
    
    for i in range(x_test.shape[0]):
        p = predict(x_test[i],w,b)
        y_pred.append(p)
        
    y_pred = np.array(y_pred)
    y_pred = y_pred.reshape(-1,1)
    print(y_pred.shape)
    print(y_test.shape)
    
    for i in range (y_test.shape[0]):
        if (y_pred[i][0] == y_test[i][0]):
            sum1 += 1
    return sum1/y_test.shape[0]
            
        
```

```python
X = pd.read_csv("./Training Data (3)/Logistic_X_Train.csv")
Y = pd.read_csv("./Training Data (3)/Logistic_Y_Train.csv")
X_Test = pd.read_csv("./Test Cases (3)/Logistic_X_Test.csv")
X_Test = X_Test.values

X = X.values
Y = Y.values

loss= []
acc = []
W = np.random.random((X.shape[1],))
b = np.random.random()

for i in range (100):
    l,w,b = grad_descent(X,Y,W,b, learning_rate = 0.5)
    acc_t= get_acc(X,Y,W,b)
    acc.append(acc_t)
    loss.append(l)

plt.plot(loss)
plt.xlabel("Time")
plt.ylabel("Negative of Log Liklihood")
plt.show()

```
