```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

dfx = pd.read_csv('Jupyter/TrainingData/Diabetes_XTrain.csv')
dfy = pd.read_csv('Jupyter/TrainingData/Diabetes_YTrain.csv')

X = (dfx.values)
print(X.shape)

Y = (dfy.values).reshape(-1,)
print(Y.shape)

l1 = (dfx.columns)
print(type(l1))
print(l1.shape)



# for i in range (X.shape[0]):
#     plt.bar(l1,X[0,:])
#     plt.show()

split= int (0.80 * X.shape[0])

X_train , Y_train = X[:split, :], Y[:split]
print(X_train.shape , Y_train.shape)

X_test , Y_test = X[split:], Y[split:]
print(X_test.shape ,Y_test.shape)

def distance(x1 , x2):
    return np.sqrt(sum((x1-x2)**2))
             
def knn(X ,Y,querypoint,k):
    vals = []
    m = X.shape[0]
             
    for i in range(m):
        d = distance(querypoint , X[i])
        vals.append((d,Y[i]))
    
    vals = sorted(vals)
    vals = vals[:k]
             
    vals = np.array(vals)
    new_vals = np.unique(vals[:,1],return_counts = True)
    index = new_vals[1].argmax()
    pred = new_vals[0][index]
    return pred
    
m = X_test.shape[0]
print(m)
for i in range (m):
    outcome = knn(X_train ,Y_train, X_test[i],100)
    print(int(outcome), Y_test[i])
    



```
