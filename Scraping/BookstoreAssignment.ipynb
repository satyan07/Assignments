```python
import bs4
from bs4 import BeautifulSoup as soup
import requests

with open ('bookstore4.csv','w') as file:
    urls = []
    for i in range (1,51):
        urls.append("http://books.toscrape.com/catalogue/page-{}.html".format(i))



        for url in urls:
            response = requests.get(url)
            response_data = soup(response.content.decode('UTF-8'), 'html.parser')
            book_element = response_data.findAll()
            img_element = response_data.findAll('div',{'class':'image_container'})
            name_element = response_data.findAll('h3',{})
            price_element = response_data.findAll('div', {'class':'product_price'})
            avail_element = response_data.findAll('p',{'class': 'instock availability'})


            for i in range(0,20):
                img_element1 = img_element[i].img.attrs['src']
                name_element1 = name_element[i].a.text
                price1 = price_element[i].findAll('p',{'class':'price_color'})
                price_final1 = price1[0].text
                avail1 = avail_element[i].text
                final_str = img_element1 + " "+ name_element1 + " " +price_final1+" "+ avail1
                final = (' '.join(final_str.split())) + '\n'
                file.write(final)
                print('c')
    #             list1 = [img_element1,name_element1 ,price_final1,avail1]
    #             file.write(str(list1))






```
